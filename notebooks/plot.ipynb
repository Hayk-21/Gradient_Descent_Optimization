```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import sys
sys.path.append("../src") #Access to the "src" directory
from optimization import GradientDescentOptimization
```

```python
# Define the function to be optimized (x0^2 + x1^2)
def function(x):
    return x[0]**2 + x[1]**2
```

```python
# Run optimization
gdo = GradientDescentOptimization(function)
result = gdo.optimize()

# Get tails
tails = np.array(result.tails)
```

```python
# 2D Plot of function(x)
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)
X, Y = np.meshgrid(x, y)
Z = function(np.array([X, Y]))
plt.contourf(X, Y, Z, cmap='viridis')

# Plot tails
plt.scatter(tails[:, 0], tails[:, 1], c='red', marker='o')

# Show plot
plt.show()
```

```python
# 3D Plot of function(x)
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)
X, Y = np.meshgrid(x, y)
Z = function([X, Y])
ax.plot_surface(X, Y, Z)

# Plot tails
ax.scatter(tails[:, 0], tails[:, 1], tails[:,2], c='red', marker='o')

# Show plot
plt.show()
```
